### Manipulação de Arquivos


**Atributos**



```python
# Criar um Arquivo

open('cidades.csv','x')
```

**Escrever no arquivo**

- parametro w - primeira escrita (cuidado apagar tudo e escreve por cima)
- parametro a - escreve no final do conteúdo que já esta no arquivo

```python
escrever = open('cidades.csv', 'w')
escrever.write('SP; Barueri; R$ 500,00\n')
escrever.write('MG; Belo Horizonte; R$781,00\n')
```

```python
escrever = open('cidades.csv', 'a')
escrever.write('BA; Salvador; R$ 86,00\n')
```

```python
# escrever no arquivo, lista inteiras

addCidades = [
     'SP; Campinas; R$ 745,00\n',
     'SP; Santana de Parnaiba; R$ 56.000,00\n',
     'MG; Monte Verde; R$ 741,00\n',
     'SE; Aracaju; R$ 1.890,00\n',
     'SP; Sorocaba; R$ 85,00\n',
     'SP; Osasco; R$ 1,99\n'
]

gravar = open('cidades.csv' , 'a')
gravar.writelines(addCidades)
```

```python
# carregar, ler o conteúdo de um  arquivo

carregar = open('cidades.csv', 'r')
mostrar = carregar.read()
carregar.close

print(mostrar)
```

```python
# apagar conteúdo do arquivo

with open('cidades.csv', 'r') as dados:
    linhas = dados.readlines()

with open('cidades.csv', 'w') as dados:
    dados.writelines(linhas[:-4])
```

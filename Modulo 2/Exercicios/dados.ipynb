```python
# Criar um Data Frame com os dados de vendas

import pandas as pd

df = pd.read_csv('dados.csv')

print(df)
```

```python
print(df.head())
```

```python
# Ver as Colulas  do banco de dados

print(df.columns)

# ver os  tipos de dados das colulas
print(df.dtypes)
```

```python
# Calcular o faturamento das vendas ( somar coluna valor)

faturamento = df['valor'].sum()

print(f'Faturamento no período foi de: R$ {faturamento:.2f}')

#calcular o faturamento medío
faturamentoMedio = df['valor'].mean()
print(f'Faturamento médio de vendas foi de: R$ {faturamentoMedio:.2f}')

```

```python
# ver lista de vendedores
print(df['vendedor'].unique())
print()
# somar as vendas por vendedor
faturamentoVendedor = df.groupby('vendedor')['valor'].sum()

print(faturamentoVendedor.sort_values(ascending=False))
```

```python
# Filtro de vendas com valores menores que quinhentos reais
faturamentoBaixo = df[ df['valor'] < 500 ]

print(faturamentoBaixo.sort_values(by='valor', ascending=False))
```

```python
# Criar um gráfico de colunas de dias e valores

import matplotlib.pyplot as plt

x = df['dias']
y = df['valor']

plt.figure(figsize=(4,2))
plt.title('Valor de Verndas por Dia')
plt.bar(x,y , color="#1696C9CA")
plt.xlabel('Dia de Vendas')
plt.xticks(x)
plt.show()
```

```python
# Crie um gráfico de barras do faturamento
# das vendas por vendedores

# Tipo coluna: bar
# tipo de barras: barh

x = df['vendedor']
y = df['valor']

plt.figure(figsize=(4,2))
plt.title('Valor de Verndas por Dia')
plt.barh(x,y , color="#015497FF")
plt.ylabel('vendedores')
plt.xlabel('Faturamento das Vendas')
#plt.xticks(y)
plt.show()
```

```python
vendedores = faturamentoVendedor.index
valores = faturamentoVendedor.values

plt.figure(figsize=(8,5))
plt.title('Valor de Verndas por Dia')
plt.pie(valores,labels=vendedores)
plt.ylabel('vendedores')
plt.xlabel('Faturamento das Vendas')
#plt.xticks(y)
plt.show()
```
